```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

## Data loading

### Subtask:
Load the data from the Excel file.


**Reasoning**:
I need to load the data from the provided Excel file into a pandas DataFrame.



```
import pandas as pd

try:
    df = pd.read_excel('IOM_DTM_ETT_SOM_Tracker_sinceFeb2025_w4.xlsx')
    display(df.head())
except FileNotFoundError:
    print("Error: 'IOM_DTM_ETT_SOM_Tracker_sinceFeb2025_w4.xlsx' not found.")
except Exception as e:
    print(f"An error occurred while reading the file: {e}")
```

## Data exploration

### Subtask:
Explore the "Reason of Displacement" column to understand its characteristics.


**Reasoning**:
Check the data type, missing values, unique entries, and frequencies for the "Reason of Displacement" column.



```
# Check data type
print(f"Data type of 'Reason of Displacement': {df['Reason of Displacement'].dtype}")

# Identify missing values
missing_count = df['Reason of Displacement'].isnull().sum()
missing_percentage = (missing_count / len(df)) * 100
print(f"\nMissing values in 'Reason of Displacement': {missing_count} ({missing_percentage:.2f}%)")

# Unique entries (case-insensitive)
unique_reasons = df['Reason of Displacement'].str.lower().unique()
print("\nUnique entries in 'Reason of Displacement' (case-insensitive):")
print(unique_reasons)

# Frequency of unique entries
reason_frequencies = df['Reason of Displacement'].value_counts()
print(f"\nFrequency of each unique entry in 'Reason of Displacement':\n{reason_frequencies}")

# Frequencies of drought-related entries
drought_reasons = [reason for reason in unique_reasons if 'drought' in reason]
drought_frequencies = df['Reason of Displacement'].apply(lambda x: x.lower() if isinstance(x,str) else x).value_counts().loc[drought_reasons]

print(f"\nFrequencies of drought-related entries:\n{drought_frequencies}")
```

## Data wrangling

### Subtask:
Filter and transform the data to focus on drought-related displacements.


**Reasoning**:
Create the 'Drought_Related' column and filter the DataFrame to include only drought-related displacements.



```
import pandas as pd

# Create 'Drought_Related' column
df['Drought_Related'] = df['Reason of Displacement'].str.contains('drought', case=False, na=False)

# Filter for drought-related displacements
df_drought = df[df['Drought_Related'] == True]

# Display the first few rows of the filtered DataFrame
display(df_drought.head())
```

## Data analysis

### Subtask:
Analyze the frequency and distribution of drought-related displacements.


**Reasoning**:
Calculate the total number of drought-related displacements, analyze the distribution across regions, and analyze the temporal distribution if a date column is available.



```
# Calculate the total number of drought-related displacements
total_drought_displacements = len(df_drought)
print(f"Total drought-related displacements: {total_drought_displacements}")

# Analyze the distribution of displacements across regions
drought_by_region = df_drought.groupby('Region Name')['Region Name'].count()
print(f"\nDistribution of drought displacements across regions:\n{drought_by_region}")

# Analyze temporal distribution if 'Date of Assessment' is available
if 'Date of Assessment' in df_drought.columns:
    try:
        df_drought['Date of Assessment'] = pd.to_datetime(df_drought['Date of Assessment'])
        df_drought['Year'] = df_drought['Date of Assessment'].dt.year
        df_drought['Month'] = df_drought['Date of Assessment'].dt.month

        drought_by_year = df_drought.groupby('Year')['Year'].count()
        print(f"\nDistribution of drought displacements by year:\n{drought_by_year}")

        drought_by_month = df_drought.groupby('Month')['Month'].count()
        print(f"\nDistribution of drought displacements by month:\n{drought_by_month}")

    except ValueError as e:
        print(f"\nError converting 'Date of Assessment' to datetime: {e}")
        print("Skipping temporal analysis due to date format issues.")

else:
    print("\n'Date of Assessment' column not found. Skipping temporal analysis.")
```
